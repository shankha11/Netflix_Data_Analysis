## DataFrame into MySQL Database

```python
#importing libraries
from sqlalchemy.exc import SQLAlchemyError
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
import pymysql
```

```python
#loading dataset
netflix_data=pd.read_csv('netflix_data_and_image/netflix_titles.csv')
```

```python
netflix_data.head()
```

```python
# MySQL database connection details
db_user = 'root'
db_password = '123456'
db_host = 'localhost'
db_name = 'netflix_db'

# Create the connection string using SQLAlchemy
connection_string = f'mysql+pymysql://{db_user}:{db_password}@{db_host}/{db_name}'

# Create an SQLAlchemy engine
try:
    engine = create_engine(connection_string)
    
    # Upload DataFrame to MySQL using SQLAlchemy engine
    netflix_data.to_sql(name='netflix_titles', con=engine, if_exists='append', index=False)
    print("DataFrame uploaded successfully to MySQL database")

except SQLAlchemyError as e:
    print(f"Error: {e}")

finally:
    # Close the connection
    engine.dispose()
    print("MySQL connection is closed")
```
